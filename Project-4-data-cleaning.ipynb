```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('books.csv')
df
```

```python
#there are a few rows that aren't formatted correctly. Any row with data in the delete column needs to be deleted
df.value_counts("delete")
```

```python
#Drop the rows where the delete column is not null
df = df[df["delete"].isnull()]
df
```

```python
#trust, but verify
df.value_counts("delete")
```

```python
#no longer need the delete column, get rid of it
df = df.drop(columns=['delete'])
df
```

```python
df.value_counts("language_code")
```

```python
#drop the irrelevant columns isbn, isbn13, and publisher
df = df.drop(columns=["isbn","isbn13","publisher"])
```

```python
#trust, but verify
df
```

```python
#change all en-US, en-GB, and en-CA to eng, then delete all non eng values
df.loc[df["language_code"] == "en-US", "language_code"] = "eng"
df.loc[df["language_code"] == "en-CA", "language_code"] = "eng"
df.loc[df["language_code"] == "en-GB", "language_code"] = "eng"

df.value_counts("language_code")
```

```python
#drop all rows where the language_code is not eng
df = df[df["language_code"] == "eng"]
df
```

```python
#trust, but verify
df.value_counts("language_code")
```

```python
#drop the language_code column since they are all the same values now
df = df.drop(columns=["language_code"])
df
```

```python
df.to_csv("cleaned_data.csv")
```
