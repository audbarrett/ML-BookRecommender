```python
import pandas as pd
df = pd.read_csv("cleaned_data.csv")
df
```

```python
# Create an empty dictionary to store author-book mappings
author_books_dict = {}

# Iterate through each row of the DataFrame
for index, row in df.iterrows():
    authors = row['authors'].split('/')  # Split authors by "/"
    book = row['title']
    
    # Iterate through each author
    for author in authors:
        author = author.strip()  # Remove leading/trailing whitespaces
        
        # Check if the author already exists in the dictionary
        if author in author_books_dict:
            # If yes, append the book to the existing list of books
            author_books_dict[author].append(book)
        else:
            # If no, create a new list with the book and assign it to the author
            author_books_dict[author] = [book]
```

```python
#flask app to connect the dictionary to the website

from flask import Flask, render_template, request

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/search', methods=['POST'])
def search():
    author = request.form['author']
    if author in author_books_dict:
        books = author_books_dict[author]
    else:
        books = []
    return render_template('result.html', author=author, books=books)

if __name__ == '__main__':
    app.run(debug=True)
```
